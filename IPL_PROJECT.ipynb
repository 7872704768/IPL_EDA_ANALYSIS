## Exploratory Data Analysis :-

**Exploratory Data Analysis(EDA)** is an **approach to analyzing data sets** to **summarize** their **main characteristics**, often with **visual methods**. 

### Q1. How many matches were played each season?

````python
no_match_season = mat_df.season.value_counts().reset_index().sort_values(by = 'season')
no_match_season
````

````python
no_match_season.T
````

````python
no_match_season.reset_index(drop = True,inplace = True)
````

````python
no_match_season
````

````python
# Create a bar plot
plt.figure(figsize=(8,4), facecolor='black')
ax = sns.barplot(x='season', y='count',data = no_match_season,palette='Reds_r')

# Set plot title and axis labels
plt.title('Number of matches played each season', fontsize=16, color='white')
plt.xlabel('Season', fontsize=14, color='white')
plt.ylabel('Counts', fontsize=14, color='white')

# Rotate x-axis labels for better readability
plt.xticks(rotation=75, color='white')
plt.yticks(ha='right', color='white')

# Set background color
plt.gca().set_facecolor('black')

# Add data labels to each bar
for i in range(len(no_match_season)):
    count = no_match_season['count'].iloc[i]
    ax.text(i, count, count, ha='center', va='bottom', color='white')

# Show the plot
plt.show()
````

- **2011-2013** have more matches being played than other seasons

- All other seasons have approximately 58-60 matches while 2011-2013 have more than 70 matches.

- Let's find out the reason for this

# Q8. Does winning toss help in winning matches?

````python
mat_df['toss_win_game_win'] = mat_df.apply(lambda row: 'Yes' if row['toss_winner'] == row['winner'] else 'No', axis=1)
mat_df.head()
````

````python
plt.figure(figsize=(9,4),facecolor='black')
ax = sns.countplot(mat_df, x="toss_win_game_win",color='darkred')
plt.title("Does Winning toss help in winning match?",fontsize=15,fontweight="bold",color='white')
plt.ylabel("Frequency", size = 20,color='white')
plt.xlabel("Winning matches based on winning toss", size = 20,color='white')
plt.xticks(size = 15,color='white')
plt.yticks(size = 15,color='white')

# Set background color
plt.gca().set_facecolor('black')

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 3, int(height), ha="center", size=15,color='white')
plt.show()
````

# Q22. How do run rates change across different phases of the innings (e.g., powerplay, middle overs, death overs) for different teams?

````python
def get_innings_phase(lst_over):
    if lst_over <= 6:
        return 'Powerplay'
    elif 7 <= lst_over <= 15:
        return 'Middle Overs'
    else:
        return 'Death Overs'
````

````python
get_innings_phase(15)
````

````python
del_df1= del_df.copy()
````

````python
del_df1['innings_phase'] = del_df1['over'].apply(get_innings_phase)
````

````python
del_df1.head(3)
````

````python
innings_phase_stats = del_df1.groupby(['batting_team', 'innings_phase']).agg(
    total_runs=('total_runs', 'sum'),
    total_overs=('innings_phase', 'count')
)
````

````python
innings_phase_stats['total_overs'] = innings_phase_stats['total_overs']//6
innings_phase_stats['run_rate'] = innings_phase_stats['total_runs'] / innings_phase_stats['total_overs']
````

````python
innings_phase_stats.reset_index(inplace=True)
innings_phase_stats
````

````python
# Define a custom color palette
custom_palette = sns.color_palette("bright", n_colors=len(innings_phase_stats['batting_team'].unique()))

plt.figure(figsize=(13, 5), facecolor='black')
sns.lineplot(data=innings_phase_stats, x="innings_phase", y="run_rate", hue="batting_team",style="batting_team", palette=custom_palette)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.title("Run Rate Trend Across Innings Phases by Batting Team",fontsize= 20, fontweight = 'bold', color='white')
plt.xlabel("Innings Phase", size = 20, color='white')
plt.ylabel("Run Rate", size = 20, color='white')
plt.xticks(size = 10, color='white')
plt.yticks(size = 14, color='white')

# Set background color
plt.gca().set_facecolor('black')
plt.show()
````

- Teams like Chennai Super Kings, Royal Challengers Bangalore maintain high run rates consistently across all phases.
- Teams like Kolkata Knight Riders and Rajasthan Royals show significant recovery in run rates during the Death Overs, highlighting their strong finishing capabilities.
